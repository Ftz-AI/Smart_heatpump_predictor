```python
#Smart Energy Management & Data Science Project

import requests
import pandas as pd

# NASA API URL and parameter setup
# Coordinates for Lagos, Nigeria
latitude = 6.5244   
longitude = 3.3792  

url = (
    f"https://power.larc.nasa.gov/api/temporal/daily/point?"
    f"parameters=ALLSKY_SFC_SW_DWN,T2M,RH2M,WS10M&"
    f"community=RE&longitude={longitude}&latitude={latitude}&"
    f"start=20240101&end=20251231&format=JSON"
)

# Requesting data from the NASA API
#Checking response status code (200 means successful request)
response = requests.get(url)


if response.status_code == 200:
    #Parsing the JSON data 
    data = response.json()

    features = data['properties']['parameter']
    df = pd.DataFrame(features)

    
    # Extracting the weather/solar features
    # Converting the 'Date' column into actual datetime format
    df.index.name = 'Date'
    df = df.reset_index()

   
    df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
    print(df.head())

   
```

```python
# ---- STEP 3: Mathematical & Thermal Modeling ----

# 1. Solar Power Generation Calculation (kWh)
# Assuming panel area = 20 sq meters, efficiency = 20% (0.2)
panel_area = 20 
efficiency = 0.2

# ALLSKY_SFC_SW_DWN is in kW-hr/m^2/day, so daily generation is calculated below:
df['Solar_Power_Generated_kWh'] = df['ALLSKY_SFC_SW_DWN'] * panel_area * efficiency


# 2. Cold Storage Thermal Cooling Load Calculation (kW)
# Assuming desired inside temperature = 4°C (Ideal for fruits/vegetables)
# U-value (insulation coefficient) = 0.3, total surface area = 50 sq meters
T_inside = 4
U_value = 0.3
surface_area = 50

# Thermal load formula: Q = U * A * (T_outside - T_inside)
# T2M represents the outside ambient temperature
df['Cooling_Load_kW'] = U_value * surface_area * (df['T2M'] - T_inside)

# Preventing negative loads (if outside temperature drops below 4°C, set to 0)
df['Cooling_Load_kW'] = df['Cooling_Load_kW'].clip(lower=0)


# Checking the data with the newly added columns (Fixed the column name casing)
print(df[['Date', 'Solar_Power_Generated_kWh', 'Cooling_Load_kW']].head())
```

```python
# ---- STEP 4: Smart Control Loop Logic ----

def control_loop(row):
    solar = row['Solar_Power_Generated_kWh']
    load = row['Cooling_Load_kW']

    
    if solar >= load:
        status = "Running on Solar & Charging Battery"
        grid_or_battery_backup_needed = 0
    
    else:
        status = "Running on Backup Power"
        grid_or_battery_backup_needed = load - solar
        
    return pd.Series([status, grid_or_battery_backup_needed])


df[['System_Status', 'Backup_Energy_Required_kWh']] = df.apply(control_loop, axis=1)
```

```python
# ---- STEP 5: Data Visualization & Plotting ----

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))


df_short = df.head(30)

# Slicing the first 30 days of data for a cleaner trend graph
plt.plot(df_short['Date'], df_short['Solar_Power_Generated_kWh'], label='Solar Power Generated (kWh)', color='orange', marker='o')
plt.plot(df_short['Date'], df_short['Cooling_Load_kW'], label='Required Cooling Load (kW)', color='blue', linestyle='--')

plt.title('Smart Cold Storage Energy Optimization (First 30 Days)')
plt.xlabel('Date')
plt.ylabel('Energy / Power')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()


print("---- FINAL PROJECT SUMMARY ----")
print(df[['Date', 'Solar_Power_Generated_kWh', 'Cooling_Load_kW', 'System_Status', 'Backup_Energy_Required_kWh']].head(10))
```
